This code will search all csv files in the direct folder with the prefix FGS and will calculate the mean of each column in each csv file. It will then return a single CSV file with means of each column and the name of the CSV page. 
it uses the glob and csv module

```python
import glob
import csv

with open('result2.csv', 'w', newline='') as f_output:
    csv_output = csv.writer(f_output)

    for filename in glob.glob('FGS*.csv'):
        print (filename)

        with open(filename, newline='') as f_input:
            csv_input = csv.reader(f_input)
            header = next(csv_input)
            averages = []

            for col in zip(*csv_input):
                averages.append(sum(float(x) for x in col) / len(col))

        csv_output.writerow([filename] + averages)
```
