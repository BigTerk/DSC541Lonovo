```python
# Import necessary packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os, glob, re, random, math, time

#noise_ratios = [5, 50, 75, 80, 90, 95]
ratio_to_readin = 5

# Define paths
PATH_TO_PNG = os.path.join(os.getcwd(), 'Distribution_Plots')
PATH_TO_CSV = os.path.join(os.getcwd(), 'Sericeability_Output')
PATH_TO_FGS = os.path.join(PATH_TO_CSV, 'FGS*_{}N*.csv'.format(ratio_to_readin))
PATH_TO_PTS = os.path.join(PATH_TO_CSV, 'PTS*_{}N*.csv'.format(ratio_to_readin))
PATH_TO_FULL = os.path.join(PATH_TO_CSV, 'FULL*_{}N*.csv'.format(ratio_to_readin))

# Get filenames of all FGS and PTS csv files
fgs_csvs = glob.glob(PATH_TO_FGS)
pts_csvs = glob.glob(PATH_TO_PTS)
full_csvs = glob.glob(PATH_TO_FULL)
```

```python
fgs_csvs
```

```python
# Loop through each file, get info, and plot Serviceability distributions - 1 per noise ratio, by FG/PT
df_fg = pd.read_csv(fgs_csvs[0])
df_fg = df_fg.drop(columns='Unnamed: 0')

# Merge all of the 5 batches of FG Serviceability into 1 dataframe
for i in range(1,len(fgs_csvs)):
    df_new = pd.read_csv(fgs_csvs[i])
    df_new = df_new.drop(columns='Unnamed: 0')
    df_fg = df_fg.merge(df_new, on='Finished Good', how='outer')

# Show df - NOTE that all columns append a letter to the end (_x, _y, _z) since the columns have the same name in each BATCH
df_fg.head()
```

```python
full_csvs
```

```python
# Loop through each file, get info, and plot Serviceability distributions - 1 per noise ratio, by FG/PT
df_full = pd.read_csv(full_csvs[0])

for i in range(1,len(pts_csvs)):
    df_new = pd.read_csv(full_csvs[i])
    df_full = df_full.merge(df_new, on=['Finished Good', 'Item_Number'], how='outer')

# Show df - NOTE that all columns append a letter to the end (_x, _y, _z) since the columns have the same name in each BATCH
df_full.head()
```

```python
pts_csvs
```

```python
# Loop through each file, get info, and plot Serviceability distributions - 1 per noise ratio, by FG/PT
df_pt = pd.read_csv(pts_csvs[0])
df_pt = df_pt.drop(columns='Unnamed: 0')

# Merge all of the 5 batches PT Serviceability into 1 dataframe
for i in range(1,len(pts_csvs)):
    df_new = pd.read_csv(pts_csvs[i])
    df_new = df_new.drop(columns='Unnamed: 0')
    df_pt = df_pt.merge(df_new, on=['Item_Number'], how='outer')

# Show df - NOTE that all columns append a letter to the end (_x, _y, _z) since the columns have the same name in each BATCH
df_pt.head()
```

```python
sns.set()
fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharex=True)


finished_goods = df_fg['Finished Good'].values
# Iterate through the five FG's
for fg in finished_goods:
    # Subset to the FG
    fg_serviceabilities = df_fg[df_fg['Finished Good'] == fg].iloc[:,1:]
    pt_serviceabilities = df_full[df_full['Finished Good'] == fg].iloc[:,2:]
    #print(serviceabilities)

    # Draw the density plot
    string = ((fgs_csvs[0].split('\\')[-1]).split('_')[2]).replace('Noise','% Inventory Level')
    fg_title_string = 'FG Serviceabilites at {}'.format(string)
    pt_title_string = 'PT Serviceabilites at {}'.format(string)
    
    sns.distplot(fg_serviceabilities.values, hist = False, kde = True, kde_kws = {'linewidth': 3}, label = fg, ax=axes[0])
    axes[0].set_title(fg_title_string, fontsize=20)
    # Plot formatting
    axes[0].legend(prop={'size': 10}, title = 'Finished Good')
    axes[0].set_xlabel('% Serviced', fontsize=16)
    axes[0].set_ylabel('Density', fontsize=16)
    #axes[0].set_ylim(0,10)
    
    sns.distplot(pt_serviceabilities.values.flatten(), hist = False, kde = True, kde_kws = {'linewidth': 3}, label = fg, ax=axes[1])
    axes[1].set_title(pt_title_string, fontsize=20)
    # Plot formatting
    axes[1].legend(prop={'size': 10}, title = 'Finished Good')
    axes[1].set_xlabel('% Serviced', fontsize=16)
    axes[1].set_ylabel('Density', fontsize=16)
    #axes[1].set_ylim(0,14)
    
plt.xlim(0, 1)
plt.tight_layout()
# output plots as images
f_name = os.path.join(PATH_TO_PNG,"{}_Inventory_DistPlots.png".format(ratio_to_readin))
fig.savefig(f_name, frameon=True)
```
